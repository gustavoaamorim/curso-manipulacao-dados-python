#### Bibliotecas

```python
import pandas as pd
import json
from google.oauth2 import service_account
```

#### Extração dos Dados

```python
# para abrir um arquivo json usamos o método open() informando o nome do arquivo e encoding como parâmetros.
# posteriormente usamos o método load() para carregar o arquivo json
f = open('tb_acoes.json', encoding='utf-8')
data = json.load(f)
```

#### Transformação dos Dados

```python
orcamentos.head()
```

```python
# realizando a junção da tabela ações com a tabela orçamentos
orcamentos = orcamentos.merge(acoes, left_on='fk_acao', right_on='id_acao', how='left')
```

```python
orcamentos
```

```python
# excluindo as colunas fk_acao e id_acao
orcamentos.drop(columns=['fk_acao', 'id_acao'], inplace=True)
```

```python
orcamentos.head()
```

```python
# fazendo a junção da tabela programas com a tabela orçamentos
orcamentos = orcamentos.merge(programas, left_on='fk_programa', right_on='id_programa', how='left')
```

```python
# exclusão das colunas fk_programa e id_programa
orcamentos.drop(columns=['fk_programa', 'id_programa'], inplace=True)
```

```python
orcamentos.head()
```

```python
# fazendo a junção da tabela unidades_orc com a tabela orçamentos 
orcamentos = orcamentos.merge(unidades_orc, left_on='fk_unidade', right_on='id_unidade', how='left')
```

```python
# excluindo as colunas fk_unidade e id_unidade
orcamentos.drop(columns=['fk_unidade', 'id_unidade'], inplace=True)
```

```python
orcamentos.head()
```

```python
# fazendo a junção da tabela anos com a tabela orçamentos
orcamentos = orcamentos.merge(anos, left_on='fk_ano', right_on='id_ano', how='left')
```

```python
# excluindo as colunas fk_ano e id_ano
orcamentos.drop(columns=['fk_ano', 'id_ano'], inplace=True)
```

```python
orcamentos.head()
```

```python
# verificando se todas as colunas possuem o mesmo número de linhas com valores
orcamentos.count()
```

```python
# verificando os tipos de dados das colunas
orcamentos.dtypes
```

```python
# o método describe() traz algumas informações estatísticas sobre os dados da tabela como contagem de linhas, média, desvio padrão, mínimo, máximo, mediana e quartis.
orcamentos[['dotacao_atual', 'empenhado', 'liquidado', 'pago']].describe()
```

```python
# verificando se não há outliers na coluna data_ano
orcamentos.data_ano.value_counts()
```

#### Carregamento dos Dados

```python
credentials = service_account.Credentials.from_service_account_file(filename='GBQ.json', scopes=['https://www.googleapis.com/auth/cloud-platform'])
```

```python
orcamentos.to_gbq(credentials=credentials, destination_table='curso_etl.etl_json', if_exists='replace')
```
